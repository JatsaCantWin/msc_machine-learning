```python
import pandas as pd

def importDataframe(dataframe_path):
    df = pd.read_csv(dataframe_path)
    return df
```

```python
def configureCategories(df):
  CategoricalColumns = ['race/ethnicity']
  Scores = ['math score', 'reading score', 'writing score', 'math score discretized', 'reading score discretized', 'writing score discretized']
  Class = 'math score'
  CategoricalColumns = [c for c in CategoricalColumns if c in df.columns]
  Scores = [s for s in Scores if s in df.columns]
  Scores = [s for s in Scores if s != Class]

  return CategoricalColumns, Scores, Class
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_bars_and_histograms(df):
    fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))

    plot_bar(axs[0, 0], df['gender'], 'Gender')
    plot_bar(axs[0, 1], df['race/ethnicity'], 'Race/Ethnicity')
    plot_bar(axs[0, 2], df['parental level of education'], 'Parental Level of Education')
    plot_bar(axs[1, 0], df['lunch'], 'Lunch')
    plot_bar(axs[1, 1], df['test preparation course'], 'Test Preparation Course')

    axs[1, 2].axis('off')

    plot_histogram(axs[2, 0], df['math score'], 'Math Score')
    plot_histogram(axs[2, 1], df['reading score'], 'Reading Score')
    plot_histogram(axs[2, 2], df['writing score'], 'Writing Score')

    plt.show()

def plot_bar(ax, data, title):
  counts = data.value_counts()
  ax.bar(counts.index, counts.values)
  ax.set_title(title)

def plot_histogram(ax, data, title):
  ax.hist(data, bins=20, alpha=0.5)
  mean = np.mean(data)
  std = np.std(data)
  ax.text(0.6, 0.8, f"{title}:\nMean: {mean:.2f}\nStd Dev: {std:.2f}")
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

def plotRelationship(df, a, b, categories):
  x = df[a]
  y = df[b]

  plt.scatter(x, y)

  plt.xlabel(a)
  plt.ylabel(b)
  plt.title('Relationship between {} and {}'.format(a, b))

  plt.show()

  categories_scores = []

  for c in categories:
    categories_scores.append(df.loc[df[a] == c, b])

  statistics, p_value = stats.ttest_ind(categories_scores[0], categories_scores[1])

  print('p-value:', p_value)
```

```python
def label_encoding(df):
    gender_map = {'male': 0, 'female': 1}
    df['gender'] = df['gender'].replace(gender_map)

    education_map = {'some high school': 0, 'high school': 1, 'some college': 2, 
                 "associate's degree": 3, "bachelor's degree": 4, "master's degree": 5}
    df['parental level of education'] = df['parental level of education'].replace(education_map)

    lunch_map = {'free/reduced': 0, 'standard': 1}
    df['lunch'] = df['lunch'].replace(lunch_map)

    test_map = {'none': 0, 'completed': 1}
    df['test preparation course'] = df['test preparation course'].replace(test_map)
```

```python
def one_hot_encode(dataframe, category):
  one_hot_encoded = pd.get_dummies(dataframe[category])
  df_encoded = pd.concat([dataframe, one_hot_encoded], axis=1)
  return df_encoded
```

```python
def remove_outliers(df):
  Q1 = df["math score"].quantile(0.25)
  Q3 = df["math score"].quantile(0.75)
  IQR = Q3 - Q1
  threshold = Q3 + 1.5*IQR

  df = df[df["math score"] < threshold]
```

```python
def descretize_result(df):
  df['math score discretized'] = df['math score'].apply(lambda x: 1 if x >= 50 else 0)
```
